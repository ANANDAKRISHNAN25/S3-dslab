```python
import numpy as np
import matplotlib.pyplot as plt
rollno = np.arange(10,10)
series1=np.array([10,12,9,14,15,9,17])
series2=np.array([12,15,14,13,18,14,12])
plt.scatter(rollno,series1)
plt.scatter(rollno,series2)
plt.legend(["series1","series2"])
plt.xlabel("rollno")
plt.ylabel("mark")
plt.show()
```
